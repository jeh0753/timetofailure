```python
import numpy as np
import pandas as pd
```

```python
def storehd5(f1,fout):
    """This will chunk by 5mil rows into a hd5 file"""
    chunksize=5000000
    data =  pd.read_csv("train.csv", sep=",", chunksize=chunksize, iterator=True, dtype={'acoustic_data': np.int16, 'time_to_failure': np.float32}, low_memory=False)                
    store = pd.HDFStore(fout, "w", complevel=9, complib="bzip2", fletcher32=True)
    i = 0 
    for df in data:
        print("working on chunk ", i)
        store.append("df", df, data_columns=True)
        i = i + 1
    store.close()     
storehd5(f1="train.csv",fout="train.hd5")
```
